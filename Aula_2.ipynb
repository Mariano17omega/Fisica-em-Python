# Gráficos

Em Matplotlib, existem duas maneiras principais de criar gráficos: usando a interface funcional e a interface orientada a objetos.



***Interface Funcional plt.plot()***

1. **Sintaxe Simples**: A interface funcional oferece uma sintaxe simples e direta para criar gráficos. Você chama diretamente funções do módulo `plt` para criar e personalizar o gráfico.

2. **Estilo Padrão**: A interface funcional segue um estilo onde você chama funções consecutivamente para criar e modificar o gráfico.

3. **Uso Global**: Uma desvantagem é que algumas configurações de gráficos podem afetar o ambiente global. Por exemplo, se você chamar `plt.xlabel()`, isso definirá o rótulo x para o gráfico atual, mesmo que você esteja criando vários gráficos.


***Interface Orientada a Objetos ax.plot()***

1. **Objetos Explícitos**: A interface orientada a objetos utiliza explicitamente objetos `Figure` e `Axes`. Você primeiro cria uma figura (`Figure`) e, em seguida, adiciona e personaliza os eixos (`Axes`) dentro dessa figura.

2. **Maior Controle**: Oferece um controle mais granular sobre o gráfico. Você pode acessar diretamente os métodos e atributos dos objetos `Axes` para personalizar o gráfico de forma detalhada.

3. **Melhor para Gráficos Complexos**: Para gráficos mais complexos com vários subplots ou personalizações detalhadas, a interface orientada a objetos é geralmente preferível.

**MRU**

```python
#MRU Usando Interface Funcional plt.plot()

import matplotlib.pyplot as plt
import numpy as np

# Constantes
x0 = 3.0  # Posição inicial (m)
v = 5.0  # Velocidade constante (m/s)

# Intervalo de tempo e duração total
tempo_total = 10.0  # Tempo total de observação (s)
instantes_de_tempo = 20
t = np.linspace(0.0, tempo_total, instantes_de_tempo, endpoint=True)
 
x = x0 + v * t
 
# Plotando o gráfico
plt.plot(t, x, label='x(t) = 3 + 5t')
plt.xlabel('Tempo (s)')
plt.ylabel('Posição (m)')
plt.title('Movimento Uniforme')
plt.legend()
plt.grid(True)

# Salvar o gráfico com alta resolução e fundo transparente
plt.savefig('MRU.png', dpi=300, bbox_inches='tight', transparent=True)

plt.show()
```

```python
# Interface Orientada a Objetos ax.plot()
import matplotlib.pyplot as plt
import numpy as np

# Constantes
x0 = 3.0  # Posição inicial (m)
v = 5.0  # Velocidade constante (m/s)

# Intervalo de tempo e duração total
tempo_total = 10.0  # Tempo total de observação (s)
instantes_de_tempo = 20
t = np.linspace(0.0, tempo_total, instantes_de_tempo, endpoint=True)
 
x = x0 + v * t

# Criando a figura e os eixos
fig, ax = plt.subplots()

# Plotando o gráfico
ax.plot(t, x, label='x(t) = 3 + 5t')
ax.set_xlabel('Tempo (s)')
ax.set_ylabel('Posição (m)')
ax.set_title('Movimento Uniforme')
ax.legend()
ax.grid(True)


# Salvar o gráfico com alta resolução
fig.savefig('MRU.png', dpi=300, bbox_inches='tight')

# Mostrando o gráfico
plt.show()
```

Parâmetros para personalizar o gráfico:

- `color`: Define a cor da linha do gráfico. Por exemplo, `'r'` para vermelho, `'b'` para azul, `'g'` para verde, etc.
- `linestyle`: Define o estilo da linha do gráfico. Por exemplo, `'-'` para uma linha contínua, `'--'` para uma linha tracejada, `':'` para uma linha pontilhada, etc.
- `linewidth` ou `lw`: Define a largura da linha do gráfico em pontos.
- `marker`: Define o tipo de marcador usado nos pontos de dados. Por exemplo, `'o'` para círculos, `'s'` para quadrados, `'x'` para x, etc.
- `markersize`: Define o tamanho dos marcadores em pontos.
- `markerfacecolor`: Define a cor do preenchimento dos marcadores.
- `markeredgecolor`: Define a cor da borda dos marcadores.
- `markeredgewidth`: Define a largura da borda dos marcadores em pontos.
 
- `alpha`: Define a transparência da linha ou dos marcadores, variando de 0 (totalmente transparente) a 1 (totalmente opaco).
- `label`: Define o rótulo da série de dados que será usado na legenda.

A lista completa de parâmetros pode ser encontrada na [documentação oficial do Matplotlib](https://matplotlib.org/stable/users/index.html). Experimente diferentes combinações para personalizar seus gráficos de acordo com suas necessidades específicas.

**Exercício 15:** Faça o grafico da Posição em função do Tempo para um oscilador harmônico simples.

Considere: $\omega = 2\pi $ $rad/s$, $A = 0.5 m$, $\phi = 0.0 rad$ e Tempo total de $5s$.


Dica: Para um oscilador harmônico simples, a força restauradora é proporcional e oposta à posição do objeto, conforme a Lei de Hooke. Matematicamente, essa relação é expressa como:

$ F = -kx $

Onde:
- $ F $ é a força restauradora,
- $ k $ é a constante da mola (rigidez da mola),
- $ x $ é a posição do objeto em relação à posição de equilíbrio.


A solução que descreve o movimento do oscilador harmônico simples é uma função do tipo:

$ 
x(t) = A \cos(\omega t + \phi) 
$

$
v(t) = -\omega A \sin(\omega t + \phi)  
$

Onde:
- $ A $ é a amplitude da oscilação,
- $ \omega $ é a frequência angular (dada por $ \omega = \sqrt{\frac{k}{m}} $),
- $ \phi $ é a fase inicial.


```python
import matplotlib.pyplot as plt
import numpy as np

# Parâmetros do movimento
w = 2.0 * np.pi # rad/2
A = 0.5  # m
phi = 0.0 #radianos
t = np.linspace(0, 5, 1000)  # array com 1000 valores de 0 a 5, em segundos.

# Calculando a posição
x = A * np.cos( w * t + phi)

# Plotando o gráfico
plt.plot(t, x, label='x(t) = 0.5 * cos(2π * t)', color = 'b',linestyle = '-.',linewidth = 2)
plt.xlabel('Tempo (s)')
plt.ylabel('Posição (m)')
plt.title('Oscilador Harmônico Simples')
plt.legend(loc='upper right') 
plt.grid(True)
plt.show()
```

**Exercício 17:** Faça 4 graficos para um oscilador harmônico simples com amortecimento (atrito):
  - Energia Total em Função do Tempo
  - Energia Cinética junto da Energia Potencial em Função do Tempo
  - Posição em Função do Tempo
  - Momento ($p=mv$) em função da posição.
     
Organize os graficos em uma figura 4x4.

Considere a massa da bola $ m = 1.0 kg$, constante da mola $ k = 10.0 N/m$, amplitude $A = 1.0 m$ e Tempo total de $25s$.
 

**Oscilador harmônico simples com amortecimento**

Para um oscilador harmônico simples com amortecimento (atrito), usamos a solução:

$ 
x(t) = A e^{-\gamma t} \cos(\omega' t) 
$

$ 
v(t) = - A e^{-\gamma t} \left( \gamma \cos(\omega' t) + \omega' \sin(\omega' t) \right) 
$

onde:
- $A $ é a amplitude inicial.
- $m $ é a massa,
- $c $ é o coeficiente de amortecimento,
- $k $ é a constante da mola,
- $x $ é a posição,
- $\gamma = \frac{c}{2m} $ é o coeficiente de amortecimento,
- $\omega' = \sqrt{\omega_0^2 - \gamma^2} $ é a frequência angular do sistema amortecido,
- $\omega_0 = \sqrt{\frac{k}{m}} $ é a frequência angular do sistema não amortecido.

  

```python
import matplotlib.pyplot as plt
import numpy as np

# Constantes
m = 1.0  # massa (kg)
k = 10.0  # constante da mola (N/m)
c = 0.2  # coeficiente de amortecimento (kg/s)
A = 1.0  # amplitude inicial (m)

# Cálculos adicionais
gamma = c / (2 * m)  # coeficiente de amortecimento
omega0 = np.sqrt(k / m)  # frequência angular não amortecida
omega_prime = np.sqrt(omega0**2 - gamma**2)  # frequência angular amortecida

# Intervalo de tempo e duração total
tempo_total = 25.0  # Tempo total de observação (s)
instantes_de_tempo = 1000
t = np.linspace(0.0, tempo_total, instantes_de_tempo, endpoint=True)

# Calculando as posições e velocidades
x = A * np.exp(-gamma * t) * np.cos(omega_prime * t)
v = -A * np.exp(-gamma * t) * (gamma * np.cos(omega_prime * t) + omega_prime * np.sin(omega_prime * t))

# Energia Potencial (Ep = 0.5 * k * x^2)
Ep = 0.5 * k * x**2

# Energia Cinética (Ek = 0.5 * m * v^2)
Ek = 0.5 * m * v**2

# Energia Total (Et = Ek + Ep)
Et = Ek + Ep

# Momento Linear (p=m*v)
p = m * v

# Criando a figura e os eixos
#fig, axs = plt.subplots(3, 1, figsize=(8, 12), sharex=True)
fig, axs = plt.subplots(2, 2, figsize=(18, 11))


# Plotando Energia Total
axs[0][0].plot(t, Et, label='Energia Total', color='black')
axs[0][0].set_ylabel('Energia Total (J)')
axs[0][0].legend()
axs[0][0].grid(True)

# Plotando Energia Cinética com a Energia Potencial
axs[0][1].plot(t, Ek, label='Energia Cinética', color='red')

axs[0][1].plot(t, Ep, label='Energia Potencial', color='blue')
axs[0][1].set_xlabel('Tempo (s)')
axs[0][1].set_ylabel('Energia (J)')
axs[0][1].legend()
axs[0][1].grid(True)

# Plotando a posição
axs[1][0].plot(t, x, label='Posição', color='y')
axs[1][0].set_xlabel('Tempo (s)')
axs[1][0].set_ylabel('Posição (m)')
axs[1][0].legend()
axs[1][0].grid(True)

# Plotando o diagrama de fase
axs[1][1].plot(x, p, label='Diagrama de fase', color='g')
axs[1][1].set_xlabel('x (m)')
axs[1][1].set_ylabel('Momento (kg.m/s)')
axs[1][1].legend()
axs[1][1].grid(True)

# Ajustando o layout
plt.suptitle('Energia em Função do Tempo para um Oscilador Harmônico Amortecido')
plt.tight_layout()
plt.show()
```

**Exercício 18:** Faça um grafico 3d para a trajetória de um movimento oblíquo.

Considere a velocidade inicial $v_0=20 m/s$, ângulo de lançamento no plano xy $\theta_{xy} = 45 ^\circ$, ângulo de lançamento em relação ao plano horizontal $\theta_{z}  = 30 ^\circ$.


Dica: A velocidade da inicial ao longo dos eixos são dadas por:

$
v_{0x} = v_0 \cos(\theta_{z} ) \cos(\theta_{xy})
$

$
v_{0y} = v_0 \cos(\theta_{z} ) \sin(\theta_{xy})
$

$
v_{0z} = v_0 \sin(\theta_{z} )
$

<img src="Velocity-vector-3D-and-components.svg" alt="Movimento oblíquo" width="300"/>


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constantes
g = 9.8  # Aceleração devido à gravidade (m/s^2)
v0 = 20.0  # Velocidade inicial (m/s)
angle_xy = 45.0  # Ângulo de lançamento no plano xy (graus)
angle_z = 30.0  # Ângulo de lançamento em relação ao plano horizontal (graus)

# Convertendo ângulos para radianos
angle_xy_rad = np.radians(angle_xy)
angle_z_rad = np.radians(angle_z)

# Componentes da velocidade inicial
v0_x = v0 * np.cos(angle_z_rad) * np.cos(angle_xy_rad)
v0_y = v0 * np.cos(angle_z_rad) * np.sin(angle_xy_rad)
v0_z = v0 * np.sin(angle_z_rad)

# Tempo total de voo
t_flight = 2 * v0_z / g



# Intervalo de tempo
t = np.linspace(0, t_flight, num=500)

# Posições
x = v0_x * t
y = v0_y * t
z = v0_z * t - 0.5 * g * t**2

# Criando a figura e os eixos 3D
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Plotando a trajetória do projétil
ax.plot(x, y, z, label='Trajetória do Projétil', color='b')

# Configurando os rótulos dos eixos
ax.set_xlabel('X (m)')
ax.set_ylabel('Y (m)')
ax.set_zlabel('Z (m)')
ax.set_title('Movimento de um Projétil em 3D')

# Adicionando uma grade e uma legenda
ax.grid(True)
ax.legend()

# Mostrando o gráfico
plt.show()
```

## Fit de Curvas

**Exercício 19:** Use os dados do experimento de MRU em exp_mru.txt e faça o ajuste de curva para obter a Posição inicial e a velocidade. Depois faça o plot das curvas.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

########################################
# Usando np.loadtxt para carregar os dados do experimeto de MRU

# Usando np.loadtxt para carregar os dados do arquivo
dados = np.loadtxt('exp_mru.txt', skiprows=1 )

# Separando os dados em duas colunas
tempo = dados[:, 0]
posicao = dados[:, 1] 
########################################

# Função modelo para MRU: x = x0 + v*t
def modelo_mru(t, x0, v):
    return x0 + v*t

# Ajuste de curva
params, covariance = curve_fit(modelo_mru, tempo, posicao)

# Parâmetros ajustados
x0, v = params

# Plotando os dados e a curva ajustada
plt.scatter(tempo, posicao, color='blue', label='Dados do Experimento')
plt.plot(tempo, modelo_mru(tempo, x0, v), color='red', label='Curva Ajustada (MRU)')
plt.xlabel('Tempo (s)')
plt.ylabel('Posição (m)')
plt.title('Ajuste de Curva para um Experimento de MRU')
plt.legend()
plt.grid(True)
plt.show()

# Exibindo os parâmetros ajustados
print("Parâmetros Ajustados:")
print("Velocidade (m/s):", v)
print("Posição Inicial (m):", x0)
```

**Exercício 20:** Use os dados do experimento de  OHS em exp_OHS.txt e faça o ajuste de curva para obter a constante da mola (k). Depois faça o plot das curvas.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

########################################
# Usando np.loadtxt para carregar os dados do experimeto de OHS
dados = np.loadtxt('exp_OHS.txt', skiprows=1)

# Separando os dados em duas colunas
posicao = dados[:, 0]
forca = dados[:, 1]
########################################

# Função modelo para um OHS: F = k * x
def modelo_ohs(x, k):
    return k * x 

# Ajuste de curva
params, covariance = curve_fit(modelo_ohs, posicao, forca)

# Parâmetro ajustado (constante elástica)
k = params[0]

# Plotando os dados e a curva ajustada
plt.scatter(posicao, forca, label='Dados do Experimento')
plt.plot(posicao, modelo_ohs(posicao, k), color='red', label='Curva Ajustada (OHS)')
plt.xlabel('Posição (m)')
plt.ylabel('Força (N)')
plt.title('Ajuste de Curva para um Oscilador Harmônico Simples')
plt.legend()
plt.grid(True)
plt.show()

# Exibindo o parâmetro ajustado (constante elástica)
print(f"Valor da Constante Elástica (k): {k :.003} N/m")
```
