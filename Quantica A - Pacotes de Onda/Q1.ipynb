```python
import scipy.integrate as integrate
import scipy
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
plt.style.use('seaborn-pastel')
```

```python
#graficos

x = np.linspace(-6,11,num=500)  #T<TC

d=2.0
A=1.0/((2*np.pi)**(1.0/4.0)*(d**(1.0/2.0)) )
X_med=2.0
P_med=2.0
h=1.0

y_x = [A*np.exp( (1.0j*t*P_med/h) -(((t-X_med)**2)/(4*d**2)) ) for t in x]
y_p = [(A*2*d/(2*h)**(1./2.))*np.exp( (1.0j*(X_med/h)*(p-P_med) ) -(((p-P_med)**2)*d**2/(h**2)) ) for p in x]

# plotting the points
axes = plt.gca()
plt.plot(x, y_x, color='blue', linestyle='-', linewidth = 2,  label='$\\left \\langle x | \\alpha \\right\\rangle $ (d=%.2f )'% d  )
plt.plot(x, y_p, color='red', linestyle='-', linewidth = 2,  label='$\\left \\langle p | \\alpha \\right\\rangle $')

plt.xlabel('$x$') # naming the x axis e naming the y axis
plt.ylabel('$\\left \\langle x | \\alpha \\right\\rangle ;  \\left \\langle p | \\alpha \\right\\rangle $') #$\\left \\langle x \\middle \\alpha \\right\\rangle $
plt.title('Pacote de Onda ($\\hbar=1$, <x>=2, <y>=2) ') # giving a title to my graph

plt.legend( loc=2)

axes.xaxis.label.set_size(18)
axes.yaxis.label.set_size(18)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.grid()


plt.savefig("Pacote de Onda.png", format="PNG",dpi=600 , bbox_inches='tight')#, pad_inches=0.5)
plt.show() # function to show the plot
```
