```python
import scipy.integrate as integrate
import scipy
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
plt.style.use('seaborn-pastel')
```

```python
#graficos

x = np.linspace(-11,11,num=1000)  #T<TC
p = np.linspace(-11,11,num=1000)  #T<TC


Delta=5.0
A=Delta/np.pi
h=1.0
A2=(2.0*np.pi*h)**(-1.0/2.0)

y_x = [A/(Delta**2+t**2) for t in x]
y_p = [A2*np.exp(-Delta*math.fabs(s)/h) for s in p]

#y_p = [(A*2*d/(2*h)**(1./2.))*np.exp( (1.0j*(X_med/h)*(p-P_med) ) -(((p-P_med)**2)*d**2/(h**2)) ) for p in x]
# plotting the points
axes = plt.gca()
plt.plot(x, y_x, color='blue', linestyle='-', linewidth = 2,  label='$\\left \\langle x | \\alpha \\right\\rangle $($\delta$=%.2f )'% Delta  )
plt.plot(p, y_p, color='red', linestyle='-', linewidth = 2,  label='$\\left \\langle p | \\alpha \\right\\rangle $')

plt.xlabel('$x;p$') # naming the x axis e naming the y axis
plt.ylabel('$\\left \\langle x | \\alpha \\right\\rangle ;  \\left \\langle p | \\alpha \\right\\rangle $') #$\\left \\langle x \\middle \\alpha \\right\\rangle $
plt.title('Lorentziana') # giving a title to my graph
plt.legend( loc=2)
axes.xaxis.label.set_size(18)
axes.yaxis.label.set_size(18)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.grid()
plt.savefig("Lorentziana 9.png", format="PNG",dpi=600 , bbox_inches='tight')#, pad_inches=0.5)
plt.show() # function to show the plot
```

```python
Delta=0.009

h=1.0


fig = plt.figure(figsize=(12,12))

ax = plt.axes(xlim=(-19,21),ylim=(-0.2,1))
plt.title('Lorentziana')
plt.rcParams.update({'font.size':20})

#axes.xaxis.label.set_size(22)
#axes.yaxis.label.set_size(22)
#plt.xticks(fontsize=22)
#plt.yticks(fontsize=22)
plt.xlabel('$x;p$') # naming the x axis e naming the y axis
plt.ylabel('$\\left \\langle x | \\alpha \\right\\rangle ;  \\left \\langle p | \\alpha \\right\\rangle $')

plt.grid()



line, = ax.plot([],[],lw=3, color='blue', linestyle='-' )
line2, = ax.plot([],[],lw=3, color='red', linestyle='-' )


def init():
    line.set_data([],[])
    return line,


def animate(i):
    s=Delta*i
    x=np.linspace(-19,21,1000)
    p=np.linspace(-19,21,1000)
    y = (s/np.pi)/(s**2+x**2)             # A*np.exp( (1.0j*x*P_med/h) -(((x-X_med)**2)/(4*(d*i)**2)) )
    y_p = A2*np.exp(-s*np.absolute(p)/h)     #(A*2*(d*i)/(2*h)**(1./2.))*np.exp( (1.0j*(X_med/h)*(p-P_med) ) -(((p-P_med)**2)*(d*i)**2/(h**2)) )



    plt.legend( ['$\\left \\langle x | \\alpha \\right\\rangle $ ($\delta$=%.2f )'% s,'$\\left \\langle p | \\alpha \\right\\rangle $'],loc=2)
    line.set_data(x,y)
    line2.set_data(p,y_p)
    return line, line2,

anim=FuncAnimation(fig, animate, init_func=init,frames=600,interval=20,blit=True)

anim.save('Lorentziana.mp4',writer='ffmpeg',fps=30)
```
